```python
import matplotlib.pyplot as plt
import numpy as np

# logs data
length_1000 = {
    "Mixed": [0.7413349763519288, 0.7797518139685073, 0.7688205168385664],
    "Max": [0.8637119636446522, 0.8602591315744488, 0.8537149552681065],
    "L1": [0.5, 0.6216839872167454, 0.7704795619883373],
    "Expectation": [0.864691364180295, 0.8939926539024067, 0.8646747440499933],
    "Average": [0.7952067544209547, 0.8483656476152487, 0.8595486210040458]
}

length_500 = {
    "Mixed": [0.7140586662141118, 0.7782772064710096, 0.769602776680653],
    "Max": [0.8145370709349121, 0.8245195213124323, 0.8234987943291261],
    "L1": [0.7453711175815958, 0.7650931200143698, 0.7574409480790986],
    "Expectation": [0.8272447646736611, 0.8531971236523044, 0.8531209445566356],
    "Average": [0.8193801660543909, 0.8093132990330576, 0.8123555057511936]
}

k_mer_values = [6, 10, 14]
pooling_layers = ["Mixed", "Max", "L1", "Expectation", "Average"]

def plot_graph_with_legend(length_data, length_label):
    x = np.arange(len(k_mer_values))
    width = 0.15

    fig, ax = plt.subplots()
    for i, (layer, values) in enumerate(length_data.items()):
        ax.bar(x + i*width, values, width, label=layer)


    ax.set_xlabel('K-mer values')
    ax.set_ylabel('AUROC on test dataset')
    ax.set_title(f'AUROC by pooling layer and K-mer values for length {length_label}')
    ax.set_xticks(x + width * 2)
    ax.set_xticklabels(k_mer_values)


    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3)

    fig.tight_layout()

    plt.show()


def plot_best_graph_no_legend_orange(length_data, length_label):
    fig, ax = plt.subplots()
    pooling_layers = list(length_data.keys())
    values = list(length_data.values())
    x = np.arange(len(pooling_layers))

    bars = ax.bar(x, values, width=0.5, color='orange')


    ax.set_xlabel('Pooling Layer (k-mer)')
    ax.set_ylabel('AUROC on test dataset')
    ax.set_title(f'AUROC by pooling layer for length {length_label}')
    ax.set_xticks(x)
    ax.set_xticklabels(pooling_layers)

    fig.tight_layout()

    plt.show()

# best k-mers
best_length_1000_with_kmer = {
    "Mixed (10)": 0.7797518139685073,
    "Max (10)": 0.8602591315744488,
    "L1 (14)": 0.7704795619883373,
    "Expectation (10)": 0.8939926539024067,
    "Average (14)": 0.8595486210040458
}

best_length_500_with_kmer = {
    "Mixed (10)": 0.7782772064710096,
    "Max (10)": 0.8245195213124323,
    "L1 (10)": 0.7650931200143698,
    "Expectation (10)": 0.8531971236523044,
    "Average (6)": 0.8193801660543909
}


plot_graph_with_legend(length_1000, 1000)
plot_graph_with_legend(length_500, 500)


plot_best_graph_no_legend_orange(best_length_1000_with_kmer, 1000)
plot_best_graph_no_legend_orange(best_length_500_with_kmer, 500)
```
